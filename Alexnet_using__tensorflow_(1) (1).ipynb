```python
import tensorflow as tf
import numpy as np

cost_history = []
n_epochs = 5

# Define your optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

# Initialize variables (not needed in most cases)
# Note: TensorFlow 2.x automatically initializes variables
# tf.compat.v1.global_variables_initializer()

train_y = np.asarray(train_y)  # Make sure train_y is a NumPy array

for i in range(n_epochs):
    with tf.GradientTape() as tape:
        logits = model(train_x, training=True)  # Assuming you have a model defined

        # Define your loss function (e.g., categorical cross-entropy)
        loss_value = tf.reduce_mean(tf.keras.losses.categorical_crossentropy(train_y, logits))

    # Calculate gradients
    grads = tape.gradient(loss_value, model.trainable_variables)

    # Apply gradients to update the model's weights
    optimizer.apply_gradients(zip(grads, model.trainable_variables))

    cost_history.append(loss_value)
    print('epoch:', i, ' - ', 'cost:', loss_value.numpy())

```
